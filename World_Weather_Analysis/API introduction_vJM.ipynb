```python
from config import api_key
query_word = "apple"
```

```python
url = f"https://api.nytimes.com/svc/search/v2/articlesearch.json?q={query_word}&api-key={api_key}"
```

```python
articles = requests.get(url).json()
```

```python
articles
```

```python
type(articles)
```

```python
articles["response"]["docs"][1]["abstract"]
```

```python
articles["response"]["docs"][1]["web_url"]
```

```python
len(articles["response"]["docs"])
```

```python
abstract_list = []
web_url_list  = []

for x in articles["response"]["docs"]:
    print(x["abstract"])
    print(x["web_url"])
    abstract_list.append(x["abstract"])
    web_url_list.append(x["web_url"])
    print("-----------")
```

```python
import pandas as pd

data = pd.DataFrame({
    "abstract" : abstract_list,
    "web_url"  : web_url_list
})
```

```python
data.to_excel(f"nyt_{query_word}.xlsx")
```
